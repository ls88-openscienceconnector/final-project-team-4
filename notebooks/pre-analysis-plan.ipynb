### Abstract

This project tries to reproduce the analysis conducted by Nima Zahadat and Kartikay Avijeet on datasets that contain information regarding movies. There are two raw data sets, which are joined to produce a cleaned data set. The original analysis consists of manipulating the raw data sets and using feature engineering to produce relevant columns and metrics that can be used as predictors for movie success. In our project, we will attempt to recreate the analysis by using the metrics created in the original project, and seeing if the results obtained in the original project hold true. We will also compare new metrics and produce visualizations to see if other variables might serve as better predictors for movie success. We will assess the reliability of the data, which although is obtained from Kaggle, the original source is TMDb which does seem to be reputable.


### Data

The data was obtained from [Kaggle](https://www.kaggle.com/tmdb/tmdb-movie-metadata). There are two raw data sets, which are combined when results are being presented to produce the final, cleaned data set. The original data from Kaggle was collected by the Movie Database API, a reputable source. The collection method is based on the TMDb community, that collects and compiles data on various movies and along various metrics such as year released and genre. The file type is csv.

### Analysis

The original analysis did not use a t-statistic as the variables were generated by combining various columns and using genre weighting to produce relevant predictors. Hence, the results were produced using bar chart visualizations, and comparing the trends of the most successful movies to see which metrics seem to have the greatest role in determining movie success. Our analysis will consist of similar metrics and visualizations to compare which features are crucial predictors. We will also attempt to create some of our own metrics, and depending on the type of metric obtained, we will see if a t-stat is possible to determine if the effect of the metric is statistically significant on movie success. A possible metric could be the absolute difference in mean rating of movies containing famous actors and mean rating of movies not containing famous actors. This test stat would be computed using:

$$ \left | \frac{1}{n} \sum_{i=1}^n s_i - \frac{1}{m} \sum_{i=1}^m \bar{s}_i \right | $$

We would also try to remove some outliers that are 2.5 standard deviations above or below the mean using: 


$$O = \{ c_i \mid c_i < \mu - 2.5 \sigma \} \cup \{ c_i \mid c_i > \mu + 2.5 \sigma \}$$
